```python
import fnmatch
for file in os.listdir('C:/Users/pedro/Downloads'):
    if fnmatch.fnmatch(file, 'Tabela_NCM_Desc_Concatenada*.xlsx'):
        
        file_correct = file
        file_source = "C:/Users/pedro/Downloads"
        file_destination = 'C:/Users/pedro/PROJETO NCM PYTHON/Base Dados/'

        get_files = os.listdir(file_source)
        
        try:
            if os.path.join(file_destination,file_correct):
                os.remove(file_destination+'/'+file_correct)
        except:
            for file1 in Path(file_source).glob(file_correct):
                shutil.move(os.path.join(file_source,file_correct),file_destination)
        else:
            for file1 in Path(file_source).glob(file_correct):
                shutil.move(os.path.join(file_source,file_correct),file_destination)
            
```
